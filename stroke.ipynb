CONVERT NUMERICAL VALUES INTO CATGERICAL VALUES

```python
df.replace({'ever_married':{'No':0,'Yes':1,}},inplace=True)
df.replace({'work_type':{'Private':1,'Self-employed':2,'Govt_job':3,'children':0}},inplace=True)
df.replace({'Residence_type':{'Urban':0,'Rural':1}},inplace=True)
df.replace({'gender':{'Male':0,'Female':1}},inplace=True)
```

```python
 X_train=df.drop(['stroke'],axis=1)
```

```python
X_train
```

```python
y_train=df.stroke
```

```python
y_train
```

SPLITTING THE VALUES

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_train,y_train,test_size=0.4)
```

NAVIE BASE ALGROTHIM

```python

from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

SUPPORT VECTOR MACHINE ALGORITHM

```python
from sklearn.svm import SVC
model = SVC()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
print(model.score(X_test, y_test))
```

```python
# Confusion Matrix
from sklearn.metrics import confusion_matrix
cf_matrix = confusion_matrix(y_test,y_pred)
print(cf_matrix)
plt.figure(figsize=(4,3))
plt.title('Confusion Matrix Visualization')
sns.heatmap(cf_matrix, annot=True, fmt='', cmap='BuPu')
```
